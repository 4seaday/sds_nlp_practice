```python
import datasets
import pandas as pd
```

```python
klue_dataset = datasets.load_dataset("klue", "mrc").remove_columns(['answers', 'guid', 'is_impossible', 'question', 'question_type', 'source'])
train_dataset = pd.DataFrame(klue_dataset['train']) 
```

```python
train_dataset.head(5)
```

# TITLE & CONTEXT DATA
- context 중복 제거
- title 중복 제거

```python
using_categories = set(train_dataset.groupby('news_category').count().sort_values(by='context', ascending=False).iloc[1:6].index)
using_categories
```

```python
train_dataset = train_dataset[train_dataset['news_category'].isin(using_categories)]
```

```python
# title
title_data = train_dataset.drop_duplicates(subset=['title'])
title_data.head(5)
```

```python
# context
context_data = train_dataset.drop_duplicates(subset=['context'])
context_data.head(5)
```

# PRE-PROCESSING

```python
import re
```

```python
# title
title_processed = [re.sub('[^가-힣a-zA-Z0-9\s]', '', x) for x in title_data['title']]
title_processed[0]
```

```python
# context
context_processed = [re.sub('[^가-힣a-zA-Z0-9\s]', '', x) for x in context_data['context']]
context_processed[0]
```

# TfidfVectorizer

```python
!pip install konlpy
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Komoran
```

```python
# context
tokenizer = Komoran()
vectorizer = TfidfVectorizer(tokenizer=lambda x: tokenizer.morphs(x))
```

```python
context_x = vectorizer.fit_transform(context_processed)
context_y = context_data['news_category']
```

```python
print(vectorizer.get_feature_names())
print(context_x.shape, context_y.shape)
```

```python
# title
tokenizer = Komoran()
vectorizer = TfidfVectorizer(tokenizer=lambda x: tokenizer.morphs(x))

title_x = vectorizer.fit_transform(title_processed)
title_y = title_data['news_category']
```

```python
print(vectorizer.get_feature_names())
print(title_x.shape, title_y.shape)
```

# K-MEANS

```python
import collections

from sklearn.cluster import KMeans
```

```python
n_clusters = 5
```

## TITLE

```python
kmeans = KMeans(n_clusters=n_clusters).fit(title_x)
```

```python
title_preds = kmeans.predict(title_x)
```

```python
def calc_cluster_counts(y_data, preds):
    clusters = {i: [] for i in range(n_clusters)}
    for i, (p, y) in enumerate(zip(preds, y_data)):
        clusters[p].append(y)
    cluster_counts = {k: collections.Counter(v) for k, v in clusters.items()}
    return cluster_counts
```

```python
cluster_counts = calc_cluster_counts(title_y, title_preds)
cluster_counts
```

```python
cluster_mapping = {k: sorted(list(v.items()), key=lambda x: x[1])[-1][0] for k, v in cluster_counts.items()}
cluster_mapping
```

```python
title_final_preds = [cluster_mapping[x] for x in title_preds]
```

### EVALUATION

```python
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix
```

```python
macro_f1 = f1_score(title_y, title_final_preds, average='macro')
micro_f1 = f1_score(title_y, title_final_preds, average='micro')
acc = accuracy_score(title_y, title_final_preds)
```

```python
print(f'macro_f1: {macro_f1}')
print(f'micro_f1: {micro_f1}')
print(f'accuracy: {acc}')
```

## CONTEXT

```python
kmeans = KMeans(n_clusters=n_clusters).fit(context_x)
context_preds = kmeans.predict(context_x)
```

```python
print(f'macro_f1: {macro_f1}')
print(f'micro_f1: {micro_f1}')
print(f'accuracy: {acc}')
```

# TEXTRANK

```python
!pip install git+https://github.com/lovit/textrank.git
```

```python
from textrank import KeywordSummarizer
from tqdm import tqdm_notebook
```

```python
tokenizer = Komoran()
def tokenize(text):
  words = tokenizer.pos(text)
  words = [w for w in words if w[1] in {'NNG', 'NNP', 'NNB', 'XR', 'VA', 'VV'}]
  assert len(words) > 0
  return words
```

```python
summarizer = KeywordSummarizer(tokenize=tokenize, min_count=2, min_cooccurrence=1)
keywords = [summarizer.summarize([x], topk=20) for x in tqdm_notebook(context_processed)]
```

```python
context_processed[0]
```

```python
keywords[0]
```

## TfidfVectorizer

```python
tmp_keywords = [' '.join([x[0][0] for x in k]) for k in keywords]
```

```python
vectorizer = TfidfVectorizer()
summarized_x = vectorizer.fit_transform(tmp_keywords)
summarized_y = context_data['news_category']
```

```python
print(vectorizer.get_feature_names())
print(summarized_x.shape, summarized_y.shape)
```
